```python
print("The user have to input one upper case letter")
txt1=input("")
if len(txt1)>10:
  print("Invalid text size:")
txt2=input("")
if len(txt1)>10:
  print("Invalid text size:")


array=[[0]*10,[1]*10]

for i in range(len(txt1)):
  array[0][i]=txt1[i]
for i in range(len(txt2)):
  array[1][i]=txt2[i]

start_1=0
for i in range(len(array[0])):

  if 65<=ord(array[0][i])<=90:
    start_1=array[0][i]


start_2=0
for i in range(len(array[1])):

  if 65<=ord(array[1][i])<=90:
    start_2=array[1][i]

print("Start for the first row:",start_1)
print("Start for the second row:",start_2)

size_1=0

for i in range(len(array[0])):
  if array[0][i]!=0:
    size_1+=1

size_2=0

for i in range(len(array[1])):
  if array[1][i]!=0:
    size_2+=1

print("Size of first row:",size_1)
print("Size of second row:",size_2)

idx1=0
for i in range(len(array[0])):
  if array[0][i]==start_1:
    idx1=i

idx2=0
for i in range(len(array[1])):
  if array[1][i]==start_2:
    idx2=i
  
print("index of first:",idx1,"index of second:",idx2)  




class CirArray:
  def __init__(self,array,start_1,start_2,size_1,size_2,idx1,idx2):
    self.start1=start_1
    self.start2=start_2
    self.size1=size_1
    self.size2=size_2
    self.idx1=idx1
    self.idx2=idx2
    self.cir=[[0]*10,[1]*10]

    for i in range(len(array[0])):
      self.cir[0][i]=array[0][idx1]
      idx1=(idx1+1)%len(self.cir[0])

    for i in range(len(array[1])):
      self.cir[1][i]=array[1][idx2]
      idx2=(idx2+1)%len(self.cir[1])

  def printARR(self):
    return self.cir

  def left(self,k):
    i=((self.idx1)+k)%len(self.cir[0])
    count=0
    while count<self.size1:
      print(self.cir[0][i],end=',')
      i=(i+k)%len(self.cir[0])

      count+=1
    
  def right(self,k):
    i=((self.idx2)-k)%len(self.cir[1])
    count=0
    while count<self.size2:
      print(self.cir[1][i],end=' ')
      i=(i+k)%len(self.cir[1])

      count+=k
  
 

array=CirArray(array,start_1,start_2,size_1,size_2,idx1,idx2)
print(array.printARR())
print(array.left(1))
print(array.right(1))

```
